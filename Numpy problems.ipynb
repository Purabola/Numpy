28. Write a NumPy program to create a 10x10 matrix, in which the elements on the borders will be equal to 1, and inside 0.

```python
matrix = np.arange(1, 101).reshape((10, 10))

for row in range(0, 10):
  if row != 0 and row != 9:
    matrix[row, [0, 9]] = 1
  else:
    matrix[row, 0:10] = 1
for row in range(1, 9):
  matrix[row, 1:9] = 0

matrix
```

```python
matrix = np.arange(1, 101).reshape((10, 10))
matrix[:, [0, 9]] = 1
matrix[1:9, 1:9] = 0
matrix[0, :] = 1
matrix[9, :] = 1

matrix
```

```python
matrix = np.ones((10, 10))
matrix[1:9, 1:9] = 0
matrix
```
